Explanation:
- sys.argv is a list of command-line arguments.
- sys.argv[0] is the script name, and sys.argv[1] is the first argument passed.
- Example run:


```python
import sys

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

if __name__ == "__main__":
    try:
        # Try to get the second argument
        num = int(sys.argv[1])
        if is_prime(num):
            print(f"{num} is a prime number.")
        else:
            print(f"{num} is not a prime number.")
    except (IndexError, ValueError):
        print("Usage: python prime_check.py <integer>")
```

Explanation:
- == checks value equality.
- is checks object identity (whether two variables point to the same memory location).
- Here, a and b have the same values but are stored separately, so a is b is False.
- c is assigned directly to a, so they reference the same object.


```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)  # True (values are equal)
print(a is b)  # False (different objects in memory)
print(a is c)  # True (same object reference)
```

Explanation:
- Negative numbers are skipped using continue.
- Execution stops at zero using break.
- Non-integer values are handled with try-except.


```python
numbers = [10, -5, "hello", 3, 0, 7]

for item in numbers:
    try:
        num = int(item)  # Convert to integer
        if num < 0:
            continue  # Skip negative numbers
        if num == 0:
            print("Zero encountered. Stopping execution.")
            break
        print(f"Processing number: {num}")
    except ValueError:
        print(f"Skipping non-integer value: {item}")
```

Explanation:
- We manually track smallest and largest values by iterating through the list.
- No built-in functions like min() or max() are used.


```python
def find_min_max(lst):
    if not lst:
        return None, None
    smallest = lst[0]
    largest = lst[0]
    for num in lst[1:]:
        if num < smallest:
            smallest = num
        if num > largest:
            largest = num
    return smallest, largest

data = [12, 5, 8, 20, 3]
min_val, max_val = find_min_max(data)
print(f"Smallest: {min_val}, Largest: {max_val}")
```
